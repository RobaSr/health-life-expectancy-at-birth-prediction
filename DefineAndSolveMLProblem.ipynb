```python
# PD - removing highly correlated feats
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
thresh = 0.9
to_drop = [column for column in upper.columns if any(upper[column] > thresh)]
df_reduced = df.drop(columns=to_drop)
```

```python
# PD - Dealing with nulls
null_list = list(df.columns[df.isna().any()])
nan_col_types = df[null_list].dtypes
nan_count = np.sum(df.isnull(), axis = 0)
condition = nan_count != 0 
```

```python
# PD - Dealing with nulls
# change names
for col in null_list:
    df[col + '_na'] = df[col].isnull()
# filling in nulls
for col in null_list:
    if df[col].dtype in ['float64', 'int64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna('missing', inplace=True)
```

```python
# PD - Dealing with nulls - checking for nulls
print(df.isnull().sum()) #it worked (yay)
```

```python
# PD - One Hot Encoding(ohe)
to_encode = list(df.select_dtypes(include=['object']))
enc = OneHotEncoder(sparse_output = False, handle_unknown='ignore')
df_enc = pd.DataFrame(enc.fit_transform(df[to_encode]), index=df.index)
df_enc.columns = enc.get_feature_names_out(to_encode)
# drop cat cols
df.drop(columns = to_encode, inplace=True)
# join cols back
df = df.join(df_enc)
```

```python
# FME (Fit model and evaluate)
feat_list = list(df.loc[:, df.columns!='Healthy life expectancy at birth'])
X = df[feat_list]
y = df['Healthy life expectancy at birth']
```

```python
#FME split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 333)
```

```python
# FME building model goal is to observe bias variance tradeoff w dif models
print('Begin GBDR Implementation (this will take a few minutes)...')
#model 1 - max_depth 2
gbdr_2_model = GradientBoostingRegressor(max_depth=2)
gbdr_2_model.fit(X_train, y_train)
gbdr_2_pred = gbdr_2_model.predict(X_test)
# model 2 - max_depth 10
gbdr_10_model = GradientBoostingRegressor(max_depth=10)
gbdr_10_model.fit(X_train, y_train)
gbdr_10_pred = gbdr_10_model.predict(X_test)
# base model(pre optimize)
base_gbdr = GradientBoostingRegressor()
print('End')
```

```python
#FME using grid search to optimize best hyperparams for best_gbdr
print('starting grid search')
param_grid = {'n_estimators': [100, 200, 300],'learning_rate': [0.1, 0.05, 0.01],'max_depth': [3, 5, 7], 'min_samples_leaf': [1, 3, 5]}
grid_search = GridSearchCV(estimator= base_gbdr, param_grid=param_grid, cv=5,scoring='neg_mean_squared_error',n_jobs=-1,verbose=2)
grid_search.fit(X_train, y_train)
print('finished')
#best params results 
print("best params:", grid_search.best_params_)
print("best mae :", grid_search.best_score_)
#best model
best_gbdr = grid_search.best_estimator_
best_gbdr.fit(X_train, y_train)
best_gbdr_pred = best_gbdr.predict(X_test)
```

```python
# TM (tweak model and improve model preformance) - bias **2, variance score and mean squared error(MSE)
#gbdr2_var
gbdr_2_evs = explained_variance_score(y_test, gbdr_2_pred)
#gbdr10_var
gbdr_10_evs = explained_variance_score(y_test, gbdr_10_pred)
#best_gbdr_var
best_evs = explained_variance_score(y_test, best_gbdr_pred)

# mean(for bias calc)
y_test_mean = np.mean(y_test)
#bias**2 for gbdr_2_model
gbdr_2_bias_sq = (np.mean(gbdr_2_pred) - y_test_mean) ** 2
#bias**2 for gbdr_10_model
gbdr_10_bias_sq = (np.mean(gbdr_10_pred) - y_test_mean) ** 2
#bias**2 for best_gbdr_model
best_gbdr_bias_sq = (np.mean(best_gbdr_pred) - y_test_mean) ** 2

#gbdr_2_mse
gbdr_2_mse = mean_squared_error(y_test, gbdr_2_pred)
#gbdr_10_mse
gbdr_10_mse = mean_squared_error(y_test, gbdr_10_pred)
#best_gbdr_mse
best_gbdr_mse = mean_squared_error(y_test, best_gbdr_pred)
# val check
#bias
bias_sqrd = [gbdr_2_bias_sq, gbdr_10_bias_sq, best_gbdr_bias_sq]
#variance
var = [1 - gbdr_2_evs, 1 - gbdr_10_evs, 1 - best_evs]
# mean squared error(mse)
mse = [gbdr_2_mse, gbdr_10_mse, best_gbdr_mse]
print('In order of md 2, md 10, and then best model\n' + '\nbias_squared is' + str(bias_sqrd) +' \nvariance is'+ str(var)+' \nand mean squared error is' +str(mse))
```

```python
#PM(Plotting model)- observing bias variance tradeoff
#models
models = ['gbdr_2 (depth=2)', 'gbdr_10 (depth=10)', 'best_gbdr (custom)']
plt.figure(figsize=(10, 6))
plt.plot(models, bias_sqrd, label='Bias**2', marker='o')
plt.plot(models, var, label='Variance (1 - EVS)', marker='o')
plt.plot(models, mse, label='Mean Squared Error (MSE)', marker='o')
plt.title('Bias Variance Tradeoff in GBDR Models')
plt.xlabel('Model')
plt.ylabel('Score')
plt.show()
```
